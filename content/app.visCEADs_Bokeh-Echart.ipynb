```python
from pathlib import Path
import yaml

def read_yaml(fn):
    with open(fn, encoding="utf8") as stream:
        try:
            #print(yaml.safe_load(stream))
            out = yaml.safe_load(stream)
        except yaml.YAMLError as exc:
            print(exc)
    return out
```

## fstring

```python
def convert_further_anyway(fstring_text, debug=False):
    if "</style>" in fstring_text:
        fstring_text_CSS_block  = fstring_text.split("</style>")[0] + "</style>"
        fstring_text_to_process = fstring_text.split("</style>")[1] + "</style>"
        return fstring_text_CSS_block + eval(f'f"""{fstring_text_to_process}"""')
    else:
        fstring_text_to_process = fstring_text
        return eval(f'f"""{fstring_text_to_process}"""')
# ---------------------------------------------------------------------------------
```

## class CEADsExplorer

```python
class CEADsExplorer(param.Parameterized):
    ## Data directly initiated and load
    sel_reg_localeJ = param.Selector(objects=reg_dict_UI_joint )#reg_all [locale] )   
    sel_year_locale = param.Selector(objects=year_all [locale]) 
    sel_indicator_main      = param.Selector(objects=Indicators_Main)
    sel_indicator_secondary = param.Selector(objects=Indicators_Secondary)
    sel_indicator_tertiary  = param.Selector(objects=sel_indicators_list[locale])  
    # class atrributes: dfN; dfE; sel_reg_localeJ; sel_year_locale; sel_indicator_main   all param!
    ## Data directly initiated and load
    dfN = param.DataFrame(dfN_locale.sort_index() )  # wrapped with param
    dfE = param.DataFrame(dfE_locale.sort_index() )  # wrapped with param
    
    # define action model 
    # TBA: tabulator selection 
    #   selectable = 'checkbox', selection = []
    def translate_df_columns (self, df, d_trans=d_trans, locale=locale):
        return (df.rename (columns=d_trans[locale]))
    def widget_Tabulator_wrapper (self, df, NorE = "N"):
        if NorE == "N":
            tabulator_filters = {
                '节点': {'type': 'list', 'func': 'in', 'valuesLookup': True, 'sort': 'asc', 'multiselect': True},
                '值':  {'type': 'number', 'func': '>=', 'placeholder': '输入最小值过滤'},
            }
            TBL = pn.widgets.Tabulator (df, show_index=False, pagination='local', page_size=Tabulator_page_size, theme = "fast", 
                                    formatters=col_fmt_dict_bokeh, text_align="right",
                                    header_filters=tabulator_filters)
        elif NorE == "E":
            tabulator_filters = {
                '投入': {'type': 'list', 'func': 'in', 'valuesLookup': True, 'sort': 'asc', 'multiselect': True},
                '产出': {'type': 'list', 'func': 'in', 'valuesLookup': True, 'sort': 'asc', 'multiselect': True},
                '行业内': {'type': 'list', 'func': 'in', 'valuesLookup': True, 'multiselect': True},
                '值':  {'type': 'number', 'func': '>=', 'placeholder': '输入最小值过滤'},
            }
            TBL = pn.widgets.Tabulator (df, show_index=False, pagination='local', page_size=Tabulator_page_size, theme = "fast", 
                                    formatters=col_fmt_dict_bokeh, text_align="right",
                                    header_filters=tabulator_filters )
        return TBL
    
    def process_param_r (self, r):
        try:
            reg = r.split(':')[1]
        except:
            reg = r
        return (reg)

    def get_df (self, r, y, i=slice(None), n=slice(None),
                            ei=slice(None), eo=slice(None), NorE = "N"):
        # processing r
        reg = self.process_param_r (r)
        if NorE == "N":
            df = self.dfN.loc[(reg,y,n,i),:].reset_index()\
                         .sort_values('value', ascending=False)
        elif NorE == "E":
            df = self.dfE.loc[(reg,y,ei,eo),:].reset_index()\
                         .sort_values('value', ascending=False)
        return  df
    
    def get_df_len (self, r, y, i=slice(None), n=slice(None),
                            ei=slice(None), eo=slice(None), NorE = "N"):
        df = self.get_df(r=r,y=y,i=i,n=n,ei=ei,eo=eo,NorE=NorE)         
        return  len(df)
    
    def get_df_with_col_translated ( self, r, y, 
                                     i=slice(None), n=slice(None), ei=slice(None), eo=slice(None),
                                     actions=["TBL_N_remove_cols","translate"], NorE = "N"):
        # get datasource        
        df = self.get_df (r,y,i=i,n=n,ei=ei,eo=eo, NorE = NorE)
        # process actions
        if "TBL_N_remove_cols" in actions:
            ## clean up two columns for Tabulator
            df = df.drop(['weight','ttips'], axis=1) 
        if "TBL_E_remove_cols" in actions:
            ## clean up one column for Tabulator
            df = df.drop(['weight'], axis=1) 
        if "translate" in actions:
            df = self.translate_df_columns (df)
        return  df
        
    def get_TBL (self, r, y, i=slice(None), n=slice(None), ei=slice(None), eo=slice(None), NorE = "N"):
        if NorE == "N":
            df = self.get_df_with_col_translated (r,y,i=i,n=n,ei=ei,eo=eo, actions= ["TBL_N_remove_cols","translate"], NorE = NorE)
            TBL = self.widget_Tabulator_wrapper (df, NorE = NorE )
        elif NorE == "E":
            df = self.get_df_with_col_translated (r,y,i=i,n=n,ei=ei,eo=eo, actions= ["TBL_E_remove_cols","translate"], NorE = NorE)
            TBL = self.widget_Tabulator_wrapper (df, NorE = NorE )
        return  TBL

myExplorer = CEADsExplorer()
```

```python
# pn.bind(gen_Network_Graph, myExplorer.param.sel_reg_localeJ, myExplorer.param.sel_year_locale, myExplorer.param.sel_indicator_main)
def gen_Network_Graph (r, y, i, repulsion, gravity):
    GdfN = myExplorer.get_df ( r=r, y=y, i=i)\
           .dropna() # remove any nodes with no weight values
    GdfN.loc[:,'ttips'] = GdfN.loc[:,'ttips'].str.replace('\n', '<br/>')
    
    GdfE = myExplorer.get_df ( r=r, y=y, NorE = "E")

    # Echart_nodes
    Echart_nodes = [
        opts.GraphNode( name=row['Node'], 
                        value=row['value'],
                        symbol_size=row['weight'],
                        tooltip_opts=opts.TooltipOpts(formatter=row['ttips']),
                        itemstyle_opts=opts.ItemStyleOpts(color=d_color[locale][row['Node']]),
                        label_opts=opts.LabelOpts(color="white", font_size=8+int(20*row['weight'])/75),
                      )
    for ind, row in GdfN.reset_index().iterrows() ]
    # Echart_edges
    Echart_edges = [
        opts.GraphLink( source=row["Input"], target=row["Output"], value=row["value"], 
                        linestyle_opts=opts.LineStyleOpts(width=row["weight"], curve=0.33, opacity=0.25),                    
                      )
    for ind, row in GdfE.reset_index().iterrows() ]  
    _plot_title_ = f"产业投入产出网络\n({r}, {y}, {i})"#"Input-Output Network of Industries"
    
    Echart_Graph = ( Graph().add(
        series_name=f"({r}, {y})",
        nodes=Echart_nodes, links=Echart_edges,
        layout="force", #"circular"
        repulsion=repulsion, gravity=gravity,
        edge_length=[10,50],
        is_roam=True, is_focusnode=True, is_draggable=True,
        label_opts = opts.LabelOpts(is_show=True),
        tooltip_opts = opts.TooltipOpts(formatter='{b}: <br />{c}'),
    )
    .set_series_opts( 
    #    textstyle_opts  = opts.TextStyleOpts( font_size=36, border_color='black', border_width=12, shadow_color='white') ,
    #    itemstyle_opts  = opts.ItemStyleOpts(  border_color='black', border_width=12) ,
        label_opts = opts.LabelOpts (font_size=16, rotate=0, text_border_width=3, text_border_color='black',
                           text_shadow_color = "whitesmoke", text_shadow_blur = .1, text_shadow_offset_x = .2, text_shadow_offset_y =.2)
        
    )
    .set_global_opts( 
        title_opts = opts.TitleOpts(title=_plot_title_) 
    )
    #.render("debug.html")
    )
    return Echart_Graph
```

```python
## Getting Interactive Graph and Putting it in the Panel
Echart_Graph_inter = pn.bind( gen_Network_Graph, 
                               myExplorer.param.sel_reg_localeJ, myExplorer.param.sel_year_locale, myExplorer.param.sel_indicator_main,
                               wg_slider_repulsion, wg_slider_gravity)

## Layout: Use panel container
pn.extension ("echarts")
pn_Echart_Graph = pn.Row ( pn.pane.ECharts( Echart_Graph_inter, width=Layout_width[locale]["Echart"], height=850),
                           pn_sidebar
                         )
                  
pn_Echart_Graph
```
